```python
run_1 = deliveries[deliveries['total_runs'] == 1]
bowler_1 = run_1.groupby('bowler').aggregate('count')
max_1 = bowler_1['match_id'].max()
bowler_1_max = bowler_1['match_id'] == max_1
bowler_1[bowler_1_max == True]
```
