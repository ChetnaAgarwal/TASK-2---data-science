 Which batsman got stumped out maximum number of times?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
deliveries = pd.read_csv('deliveries.csv')
deliveries.head()
```

```python
deliveries['dismissal_kind'].unique()
```

```python
stumped_out = deliveries[deliveries['dismissal_kind'] == 'stumped']
stumped_out
```

```python
stumped_out_count = stumped_out.groupby('batsman').aggregate('count')
max = stumped_out_count['match_id'].max()
stumped_out_count
```

```python
max
```

```python
stumped_out_count[stumped_out_count['match_id'] == max]
```
