```python
import pandas as pd
import numpy as np
```

```python
deliveries = pd.read_csv('deliveries.csv')
```

```python
deliveries.head()
```

QUERY 1 - which bowler has played the minimum no of balls?

```python
count1 = deliveries.groupby('bowler').aggregate('count')
min1 = count1['match_id'].min()
count1
```

```python
min1
```

```python
bool1 = count1['match_id'] == min1
bool1
```

```python
count1[bool1 == True]
```

ANSWER 1 - AC GILCHRIST

QUERY 2 - which team has batted maximum no of times in the first inning?

```python
first_inning = deliveries[deliveries['inning'] == 1]
count2 = first_inning.groupby('batting_team').aggregate('count')
max2 = count2['match_id'].max()
count2
```

```python
bool2 = count2['match_id'] == max2
```

```python
count2[bool2 == True]
```

ANSWER 2 - MUMBAI INDIANS

```python
ven_count = matches.groupby('venue').aggregate('count')
ven_count
```
